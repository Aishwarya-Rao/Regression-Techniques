Predicting salary using Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_input=pd.read_csv("Salary_dataset.csv")
df_input.head(5)
```

```python
df_input.drop(columns='Unnamed: 0',inplace=True)
df_input.columns
```

```python
X_train=df_input["YearsExperience"]
# X_train=X_train[:25]
X_train_modi=np.array(X_train)
Y_train=df_input["Salary"]
Y_train=np.array(Y_train)
```

```python
plt.scatter(X_train,Y_train)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
lr=LinearRegression()
```

```python
lr_model=lr.fit(np.array(X_train).reshape(-1,1),Y_train)
```

```python
print(lr_model.intercept_,lr_model.coef_)
```

```python
Y_line=X_train*lr_model.coef_ + lr_model.intercept_
```

```python
plt.scatter(X_train,Y_train)
plt.plot(X_train,Y_line)
```

```python
X_test=df_input["YearsExperience"]
X_test=np.array(X_test[25:])
Y_test=df_input["Salary"]
Y_test=np.array(Y_test[25:])
print(X_test,Y_test)
```

```python
lr_model.score(X_test.reshape(-1, 1),Y_test)
```

```python
Y_pred=lr_model.predict(X_test.reshape(-1,1))
Y_pred
```

```python
r2 = r2_score(Y_test, Y_pred)
r2
```
